1. Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
nums = [1, 4, 6, 8, 10]
n = int(input("Enter a number to append: "))

#append
nums.append(n)
print("After append:", nums)

#extend
nums.extend([2, 3])
print("After extend:", nums)

# insert
nums.insert(1, 5) 
print("After insert:", nums)

#remove
nums.remove(1)  
print("After remove:", nums)

# pop
nums.pop(3) 
print("After pop:", nums)

# index
print("Index of 5:", nums.index(5))

# count
print("Count of 5:", nums.count(5))

# sort
nums.sort()
print("After sort:", nums)

# reverse
nums.reverse()
print("After reverse:", nums)

# clear
nums.clear()
print("After clear:", nums)
```

11. Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
data = {
    'Name': ['sam','sana','anu','ritu','manu','harsh','gitu','ash','sana','ritu','sam'],  
    'Age':  [25, 22, 26, 35, 29, 31, 27, 30, 22, 35, 25],          
    'City': ['A','F','L','F','N','L','S','Y','F','F','A'],       
    'Salary': [50000, 55000, 52000, 58000, 60000, np.nan, 57000, 59000, np.nan, 58000, 50000]  
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
print("DataFrame info:\n")
df.info()

print("DataFrame description:\n")
df.describe()
```

```python
print("Selected Columns (Name & City):\n")
print(df[['Name', 'City']])
```

```python
df_without_city = df.drop('City', axis=1)
print("DataFrame without City column:\n", df_without_city)
```

```python
salary_mean = df['Salary'].mean()
df['Salary'].fillna(salary_mean, inplace=True)
print("DataFrame after filling missing Salary values:\n", df)
```

```python
df.drop_duplicates(inplace=True)
print("DataFrame after removing duplicates:\n", df)
```
