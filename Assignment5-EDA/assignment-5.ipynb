# Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as  sns
```

# <font color="TEAL"> 1. Data Cleaning and Preparation:</font>

```python
#loading the dataset
data = pd.read_csv("Cardiotocographic.csv")
data
```

Handling Missing values

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
#filling missing numeric values with mean
data = data.fillna(data.mean(numeric_only=True))
```

```python
data.isnull().sum()
```

Correcting  data type inconsistencies

```python
data.dtypes
```

```python
for col in data.columns:
    if data[col].dtype=='object':
        try:
            data.col=pd.to_numeric(data[col])
        except:
            pass
```

```python
data.dtypes
```

Detect and treat outliers

```python
def remove_outliers(df):
    numeric_cols =df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        q1=df[col].quantile(0.25)
        q3=df[col].quantile(0.75)
        IQR = q3-q1
        lower =q1-1.5*IQR
        upper =q3+1.5*IQR
        df=df[(df[col]>=lower)&(df[col]<= upper)]
    return df
        
```

```python
data = remove_outliers(data)
```

2. High standard deviation or IQR:

3. Low standard deviation:

4. Columns with 0 std or IQR:

# <font color='teal'>4. Pattern Recognition and Insights:</font>

1. Identifying Correlations Between Variables

```python
# feature 0  feature relationship
corr_matrix =data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
#feature- target relationship
data['NSP_num']= data['NSP'].astype(int)
corr_matrix = data.select_dtypes(include=['number']).corr()
```

```python
print("\nTop negative correlations with NSP:\n", target_corr.tail(6))
```
