# Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset

# Descriptive Analytics for Numerical Columns

```python
import pandas as pd
```

```python
#Load the dataset
data=pd.read_csv("sales_data_with_discounts.csv")
print(data.head()) #first five rows
```

```python
#numerical columns
num=data.select_dtypes(include=['number'])
print(num)#printing numerical columns in a list
```

```python
print(data.dtypes) 
```

```python
#Calculate the mean, median, mode, and standard deviation for numerical columns
mean_num=num.mean()
print("Mean:")
print(mean_num)
median_num=num.median()
print("Median:")
print(median_num)
mode_num=num.mode().iloc[0] #selects the first element from the mode
print("Mode:")
print(mode_num)
std_num=num.std()
print("Standard Deviation:")
print(std_num)
```

```python
num.describe() # detailed info
```

```python
# Combining into one summary table
summary=pd.DataFrame({'Mean' : mean_num,
                      'Median' : median_num,
                      'Mode' : mode_num,
                      'Standard Deviation' : std_num})
print(summary)
```

# Provide a brief interpretation of these statistics.
From the descriptive Analytics, I understood the overall pattern and spread of the numerical data.
The mean and median were quite close for most columns, showing that the data is fairly balanced without many extreme outliers.
The standard deviation helped me to see how much variation there is in sales and discounts across records.

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Histogram

```python
for col in num:
    plt.figure(figsize=(10, 6))  # one large histogram per column
    sns.histplot(data[col], kde=True, bins=30, color='skyblue')
    plt.title(f"Distribution of {col}", fontsize=16)
    plt.xlabel(col, fontsize=14)
    plt.ylabel("Frequency", fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

```

Boxplot

```python
for col in num:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=data[col], color='lightgreen')
    plt.title(f"Box Plot of {col}", fontsize=16)
    plt.xlabel(col, fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

Bar Chart Analysis for Categorical Column:

```python
categorical_cols=data.select_dtypes(include=['object','category'])
print(categorical_cols)
```

```python
for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    data[col].value_counts().plot(kind='bar', color='lightcoral')
    plt.title(f"Category Distribution of {col}", fontsize=16)
    plt.xlabel(col, fontsize=14)
    plt.ylabel("Count", fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

# Summarize the key findings from the data visualization
From the histograms, I noticed how each numerical column is distribute, some were normally spread while others were skewed.
The boxplots made it easier to spot outliers and compare the range of values between columns.
For the categorical columns, the bar charts showed which categories appeared the most.
This helped me to understand which products, regions, or customer types were more common in the dataset.
Overall, the analysis gave a clear picture of data trends, patterns, and possible outliers.
